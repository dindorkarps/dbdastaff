```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Final.csv")
df
```

```python
df7=df.drop(df[df['DepartmentId'] == 1].index,axis=0,inplace=True)
df7
```

```python
# Assuming 'df' is your DataFrame and 'date_column' is the column containing the dates
df['LogDate'] = pd.to_datetime(df['LogDate'], format='%d-%m-%Y %H:%M')
df
```

```python
df.head(20)
```

```python

# Assuming 'datetime_column' is the datetime column you want to split
# Convert the datetime column to datetime type if it's not already in datetime format
df['datetime_column'] = pd.to_datetime(df['LogDate'])

# Split datetime column into date and time components
df['date'] = df['datetime_column'].dt.date
df['time'] = df['datetime_column'].dt.time
df['hour'] = df['datetime_column'].dt.hour
df['minute'] = df['datetime_column'].dt.minute
df
```

```python
df5=df.drop(['LogDate','datetime_column','hour','minute'],axis=1)
df5
```

```python
df5.head(20)
```

```python
min_time = df.groupby(['date', 'EmployeeCode'])['time'].min()
max_time = df.groupby(['date', 'EmployeeCode'])['time'].max()

# Output the result
print("Date-wise and roll number-wise minimum time:")
print(min_time)
print("\nDate-wise and roll number-wise maximum time:")
print(max_time)
```

```python
min_time
```

```python
mt=pd.DataFrame(min_time)
mt
```

```python
mn=pd.DataFrame(max_time)
mn
```

```python
merged_df = pd.merge(mt, mn, on='date', how='inner')
merged_df
```

```python
# Assuming 'df' is your DataFrame
merged_df.reset_index(inplace=True)
```
